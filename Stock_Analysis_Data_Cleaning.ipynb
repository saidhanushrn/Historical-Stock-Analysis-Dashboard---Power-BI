```python
import pandas as pd

# Load CSV
df = pd.read_csv(r"E:\sai_dhanush_data_analyst_projects\Finance\Stock_market_analytics_dashboard\RELI Historical Data.csv")

# Remove commas from numeric columns
for col in ['Price', 'Open', 'High', 'Low']:
    df[col] = df[col].str.replace(',', '').astype(float)

# Remove % sign and convert Change column to float
df['Change %'] = df['Change %'].str.replace('%', '').astype(float)

# Volume as is (clean later)
df['Vol.'] = df['Vol.'].str.replace('M', '').astype(float) * 1e6

# Save cleaned CSV for import
df.to_csv("E:\sai_dhanush_data_analyst_projects\Finance\Stock_market_analytics_dashboard\RELI Historical Data_cleaned.csv", index=False)
```

```python
import pandas as pd

# Load CSV with original date format
df = pd.read_csv(r"E:\sai_dhanush_data_analyst_projects\Finance\Stock_market_analytics_dashboard\RELI Historical Data_cleaned.csv")

# Convert date format from DD-MM-YYYY to YYYY-MM-DD
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')

# Save to a new CSV
df.to_csv("E:\sai_dhanush_data_analyst_projects\Finance\Stock_market_analytics_dashboard\RELI Historical Data_cleaned_date.csv", index=False)
```
